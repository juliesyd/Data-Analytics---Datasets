## Part 2 - Line Plot 
I want to investigate how a persons annual income might affect the number of books they read. To do this I am plotting a line graph with ranges of income on the x-axis and the average number of books read on the y-axis.


```python
df.groupby('Incomes')['How many books did you read during last 12months?'].mean().plot.line(
                                              rot = 75,
                                              xlabel = 'Annual Income',
                                              ylabel = 'Average number of books read',
                                              title = 'Annual Income and Age',
                                              marker = 'o',
                                              color = 'olive')
```

The lowest number of books read a year is by those who earn less than $10,000 a year. It can be seen that the number of books read increases as the income increases when entering the 20,000-30,000 bracket. The number of books read then fluctuates with the increase income.Surprisingly, this number declines drastically as the income enters the highest bracket in this dataset,75,000-100,000.

## Part 3 - Bar Plot 
I want to investigate the correlation between the level of education that a person has recieved and the number of books they have read. To do this, the level of education is plotted on the x axis, against the average number of books read in the past year on the y axis. This bar chart shows the average number of books read by each level of education.


```python
df.groupby('Education')['How many books did you read during last 12months?'].mean().plot.bar(title = 'Books Read per Year',
                                                                                             xlabel = 'Level of education recieved',
                                                                                             ylabel = 'Average number of books read',
                                                                                             color = 'cyan',
                                                                                             rot = '75')
```

It can be seen that the highest number of books read a year on average was done by those with some form of post-graduate degree/training, at over 20 books.The lowest amount of books read a year averages at less then 5, done by those with no form of education. In general, it seems that the higher the level of education a person has recieved, the more books they will read a year.
